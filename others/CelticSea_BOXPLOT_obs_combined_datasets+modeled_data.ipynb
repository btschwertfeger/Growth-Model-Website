##### Celtic Sea observed weith-at-age in stock and landings (from ICES)
##### Boxplots represent median (green) and interquartile range (IQR) in data distribution (over time)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D
```

```python
# Load observational datasets using pandas
obs_df1 = pd.read_csv('data/CelticSea_PREPARED-FOR-BOX-PLOTTING_7e-k_stock_weight-at-age_1971-2018_WGCSE2019.csv', index_col=0, sep=' ')
```

```python
# Check the structure of the dataset
obs_df1.tail()
```

```python
min(obs_df1['10']),max(obs_df1['10'])
```

```python
# Create a figure of particular size and set the number of axes
f, ax = plt.subplots(1, 1, dpi=300, figsize=(6, 6), facecolor='white')

# Plot observations as a boxplot
obs_df1.loc['1971':'2007'].boxplot(showfliers=False) 

ax.set_yticks(range(1, 26, 2))
ax.set_title('Celtic Sea', fontsize=14)

line1 = Line2D([0], [0], color='green', linewidth=1, linestyle='-')
line2 = Line2D([0], [0], color='blue', linewidth=1, linestyle='-')

lines = [line1, line2]
labels = ['Median', 'IQR']

plt.legend(lines, labels, 
         bbox_to_anchor=(0.01, 0.01, 1., 0.8), 
         loc='lower right', 
         ncol=3,
         fontsize=12) 

# Plot labels
plt.ylabel('Weight, kg', fontsize=10)
plt.xlabel('Age, years', fontsize=10)

# Hide grid cells
ax.grid(False)
```
